## 1. What is a Time Series?

A **time series** is a sequence of observations indexed by time:

$$
y_1, y_2, \ldots, y_T
$$

Examples:

- Daily hotel occupancy rate
- Monthly airline passengers
- Weekly sales revenue
- Hourly website visits
- Daily electricity demand
- Monthly unemployment rate

The key difference from ordinary tabular data is that **the order matters**. Observations close in time are often related.



## 2. Components of a Time Series

A common conceptual decomposition is:

$$
y_t = T_t + S_t + C_t + \varepsilon_t
$$

where:

- $T_t$ = trend component
- $S_t$ = seasonal component
- $C_t$ = cyclical component
- $\varepsilon_t$ = irregular error/noise component

For many business forecasting tasks, we often simplify this to:

$$
y_t = T_t + S_t + \varepsilon_t
$$

### Additive vs Multiplicative Decomposition

**Additive form:**

$$
y_t = T_t + S_t + \varepsilon_t
$$

Useful when seasonal variation is roughly constant over time.

**Multiplicative form:**

$$
y_t = T_t \times S_t \times \varepsilon_t
$$

Useful when seasonal variation increases as the level of the series increases.



## 3. Univariate vs Multivariate Time Series

### Univariate forecasting

A univariate model forecasts one variable using its own past values:

$$
\hat{y}_{t+h} = f(y_t, y_{t-1}, y_{t-2}, \ldots)
$$

Example: forecasting next month's hotel occupancy using past occupancy only.

### Multivariate forecasting

A multivariate model uses several related variables:

$$
\hat{y}_{t+h} = f(y_t, x_{1,t}, x_{2,t}, \ldots)
$$

Example: forecasting occupancy using past occupancy, room price, marketing spend, weather and events.



## 4. Linear vs Non-Linear Forecasting

### Linear models

Linear time series models assume the future value is a linear combination of past values and/or past errors:

$$
y_t = c + \phi_1y_{t-1} + \phi_2y_{t-2} + \cdots + \varepsilon_t
$$

Examples:

- AR
- MA
- ARMA
- ARIMA
- VAR

### Non-linear models

Non-linear models allow complex relationships:

$$
y_t = f(y_{t-1}, y_{t-2}, x_t) + \varepsilon_t
$$

where $f(\cdot)$ may be learned by a machine learning algorithm.

Examples:

- Random Forest regression
- Gradient Boosting
- Neural networks
- LSTM models

Important note: non-linear models are not automatically better. They usually require more data, careful validation and feature engineering.



## 5. Random Walk

A **random walk** assumes that the next value equals the current value plus a random shock:

$$
y_t = y_{t-1} + \varepsilon_t
$$

where $\varepsilon_t$ is white noise.

The simplest forecast from a random walk is:

$$
\hat{y}_{t+1} = y_t
$$

This is also called a **naive forecast**.

### Business interpretation

If hotel occupancy behaves like a random walk, the best forecast for tomorrow may simply be today's occupancy. This can be a strong baseline for short-term forecasting.



## 6. Exponentially Weighted Moving Average / Simple Exponential Smoothing

Simple exponential smoothing updates the forecast using a weighted average of the most recent observation and previous smoothed value:

$$
\ell_t = \alpha y_t + (1-\alpha)\ell_{t-1}
$$

where:

- $\ell_t$ = smoothed level at time $t$
- $\alpha$ = smoothing parameter, $0 < \alpha < 1$
- $y_t$ = observed value

A higher $\alpha$ gives more weight to recent observations.

Forecast:

$$
\hat{y}_{t+h} = \ell_t
$$

Simple exponential smoothing is useful when there is **no clear trend or seasonality**.



## 7. Holt's Linear Trend Method

Holt's method extends simple exponential smoothing by adding a trend component.

Level equation:

$$
\ell_t = \alpha y_t + (1-\alpha)(\ell_{t-1} + b_{t-1})
$$

Trend equation:

$$
b_t = \beta(\ell_t - \ell_{t-1}) + (1-\beta)b_{t-1}
$$

Forecast equation:

$$
\hat{y}_{t+h} = \ell_t + hb_t
$$

where:

- $\ell_t$ = level
- $b_t$ = trend
- $h$ = forecast horizon
- $\alpha, \beta$ = smoothing parameters

Useful when the time series has a trend but no seasonality.



## 8. Holt-Winters Seasonal Method

Holt-Winters extends exponential smoothing to include level, trend and seasonality.

For the additive seasonal version:

Level:

$$
\ell_t = \alpha(y_t - s_{t-m}) + (1-\alpha)(\ell_{t-1} + b_{t-1})
$$

Trend:

$$
b_t = \beta(\ell_t - \ell_{t-1}) + (1-\beta)b_{t-1}
$$

Seasonality:

$$
s_t = \gamma(y_t - \ell_t) + (1-\gamma)s_{t-m}
$$

Forecast:

$$
\hat{y}_{t+h} = \ell_t + hb_t + s_{t+h-m(k+1)}
$$

where:

- $m$ = seasonal period, e.g. 12 for monthly data with yearly seasonality
- $\gamma$ = seasonal smoothing parameter

Useful when the series has both **trend and repeated seasonal patterns**.



## 9. Stationarity

A time series is **stationary** when its statistical properties do not change over time.

A weakly stationary series has:

1. Constant mean:

$$
E(y_t) = \mu
$$

2. Constant variance:

$$
Var(y_t) = \sigma^2
$$

3. Autocovariance depends only on lag $k$, not on time $t$:

$$
Cov(y_t, y_{t-k}) = \gamma_k
$$

Many ARIMA-style models require stationarity or are built after transforming the data to become stationary.



## 10. Differencing

Differencing removes trend by subtracting consecutive observations:

$$
\Delta y_t = y_t - y_{t-1}
$$

Second-order differencing:

$$
\Delta^2 y_t = \Delta y_t - \Delta y_{t-1}
$$

Seasonal differencing:

$$
\Delta_m y_t = y_t - y_{t-m}
$$

where $m$ is the seasonal period.

In ARIMA$(p,d,q)$, $d$ represents the number of non-seasonal differences used.



## 11. Autocorrelation, ACF and PACF

### Autocorrelation

Autocorrelation measures the relationship between a series and its lagged values:

$$
\rho_k = \frac{Cov(y_t, y_{t-k})}{Var(y_t)}
$$

### ACF: Autocorrelation Function

The ACF shows correlations at different lags:

$$
\rho_1, \rho_2, \rho_3, \ldots
$$

It helps identify persistence, seasonality and moving-average structures.

### PACF: Partial Autocorrelation Function

The PACF measures the direct relationship between $y_t$ and $y_{t-k}$ after removing the effect of intermediate lags.

General rule of thumb:

| Pattern | Possible model |
|---|---|
| PACF cuts off, ACF tails off | AR model |
| ACF cuts off, PACF tails off | MA model |
| Both tail off | ARMA / ARIMA model |



## 12. AR Model

An autoregressive model of order $p$, AR$(p)$, uses past values of the series:

$$
y_t = c + \phi_1y_{t-1} + \phi_2y_{t-2} + \cdots + \phi_py_{t-p} + \varepsilon_t
$$

where:

- $p$ = number of autoregressive lags
- $\phi_i$ = autoregressive coefficients
- $\varepsilon_t$ = white noise error

AR models are useful when current values depend on previous values.



## 13. MA Model

A moving-average model of order $q$, MA$(q)$, uses past forecast errors:

$$
y_t = \mu + \varepsilon_t + \theta_1\varepsilon_{t-1} + \theta_2\varepsilon_{t-2} + \cdots + \theta_q\varepsilon_{t-q}
$$

where:

- $q$ = number of moving-average terms
- $\theta_i$ = moving-average coefficients
- $\varepsilon_t$ = white noise shock

MA models are useful when a shock affects the next few observations before fading away.



## 14. ARMA Model

ARMA combines AR and MA components:

$$
y_t = c + \phi_1y_{t-1} + \cdots + \phi_py_{t-p} + \varepsilon_t + \theta_1\varepsilon_{t-1} + \cdots + \theta_q\varepsilon_{t-q}
$$

ARMA$(p,q)$ is appropriate for **stationary** time series.



## 15. ARIMA Model

ARIMA stands for:

- **AR**: Autoregressive part
- **I**: Integrated part, meaning differencing
- **MA**: Moving-average part

ARIMA$(p,d,q)$ means:

- $p$ = number of AR lags
- $d$ = number of differences
- $q$ = number of MA terms

The model is applied to the differenced series:

$$
\Delta^d y_t
$$

When $d=0$, ARIMA becomes ARMA.



## 16. Box-Jenkins Methodology

The Box-Jenkins approach is a structured process for ARIMA modelling.

### Step 1: Identification

- Plot the series
- Check stationarity
- Difference if needed
- Inspect ACF and PACF
- Select candidate $p$, $d$, $q$

### Step 2: Estimation

- Fit candidate ARIMA models
- Estimate parameters
- Compare AIC/BIC

### Step 3: Diagnostic Checking

- Residuals should behave like white noise
- Check residual ACF
- Check Ljung-Box test
- Inspect normality if prediction intervals matter

### Step 4: Forecasting

- Generate forecasts
- Compare against holdout data
- Translate errors into business decisions



## 17. Forecast Evaluation Metrics

Let $y_t$ be the actual value and $\hat{y}_t$ the forecast.

### Error

$$
e_t = y_t - \hat{y}_t
$$

### Mean Absolute Error

$$
MAE = \frac{1}{n}\sum_{t=1}^{n}|y_t - \hat{y}_t|
$$

### Mean Squared Error

$$
MSE = \frac{1}{n}\sum_{t=1}^{n}(y_t - \hat{y}_t)^2
$$

### Root Mean Squared Error

$$
RMSE = \sqrt{\frac{1}{n}\sum_{t=1}^{n}(y_t - \hat{y}_t)^2}
$$

### Mean Absolute Percentage Error

$$
MAPE = \frac{100}{n}\sum_{t=1}^{n}\left|\frac{y_t - \hat{y}_t}{y_t}\right|
$$

### Warning about MAPE

MAPE can be misleading when actual values are close to zero. In those cases, MAE or RMSE may be safer.



## 18. Multivariate Time Series: VAR

A Vector Autoregression model uses several variables and their lags.

For two variables, $y_t$ and $x_t$:

$$
y_t = c_1 + a_{11}y_{t-1} + a_{12}x_{t-1} + \varepsilon_{1,t}
$$

$$
x_t = c_2 + a_{21}y_{t-1} + a_{22}x_{t-1} + \varepsilon_{2,t}
$$

VAR is useful when variables influence each other over time.

Example: hotel occupancy may influence ADR, and ADR may also influence future occupancy.



## 19. Feature-Based Forecasting for Machine Learning

To use machine learning for time series, we convert temporal structure into features.

Typical features:

$$
y_{t-1}, y_{t-2}, y_{t-7}, y_{t-12}
$$

Rolling statistics:

$$
RollingMean_t = \frac{1}{k}\sum_{i=1}^{k}y_{t-i}
$$

Calendar features:

- day of week
- month
- holiday flag
- high-season flag

External variables:

- marketing spend
- room price
- weather
- events

Machine learning models can capture non-linear relationships but still require careful time-aware validation.



## 20. Practical Interpretation Checklist

When evaluating a forecasting model, ask:

1. Does it beat a naive forecast?
2. Are the residuals random?
3. Is the model interpretable enough for the business context?
4. Is the forecast horizon realistic?
5. Are prediction intervals required?
6. What decision will the forecast support?

A forecasting model should not only be statistically acceptable. It must also be useful for decision-making.
